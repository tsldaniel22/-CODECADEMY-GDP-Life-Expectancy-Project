## Importing of Python modules/libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import seaborn as sns
import statsmodels





```

## Brief examining of data (first 5 rows)

```python
data = pd.read_csv('all_data.csv')
```

```python
print(data.head(5))
```

```python
data = data.rename(columns = {'Country':'country', 'Year':'year', 'Life expectancy at birth (years)': 'life expectancy', 'GDP':'GDP'})

```

```python
data = data.rename(columns = {'country':'country', 'year':'year', 'lifeexpectancy': 'life expectancy', 'GDP':'GDP'})
print(data.head(5))
data.columns = data.columns.str.replace(" ", "_")
print(data.head(5))
print(data.country.unique())

np.mean(data.life_expectancy)
np.max(data.year)
```

The average life expectancy of the six countries rounds off to about 72.8, which is a considerably high number. This could likely be due to the presence of developed countries such as China, Germany and USA where residents are generally able to afford a higher standard of living, keeping them away from chronic illnesses and diseases, and thus resulting in generally high life expectancies in these nations. 

## Exploring the data through visualizations
###Let's first take a look at the distribution of life expectancy and GDP across the six countries. 

```python
plt.subplot(1,2,1)
plt.hist(data.life_expectancy, bins = 20)
plt.title("Life Expectancy distribution")
plt.xlabel("Life expectancy")
plt.ylabel("Tally")
plt.xticks([40,50,60,70,80])


plt.subplot(1,2,2)
plt.hist(data.GDP)
plt.title("GDP distribution")
plt.xlabel("GDP")
plt.ylabel("Tally")

plt.subplots_adjust(wspace = 0.7)
```

Based on the two histograms plotted above, we can see that the distribution of life expectancy is left-skewed (as its tail extends to the left) while the distrubution of GDP is right-skewed (as its tail extends to the right).

Taking a closer look at the distribution of **life expectancy** , we can infer that there is generally a high level of life expectancy across the 6 countries, as evident in the relatively higher numbers in the age range of 70-80 as compared to a lower range of say, 50-60.

This is as opposed to the generally low levels of GDP across the 6 counntries, as seen by the much higher count in a low GDP range of 0.0 to 0.5 as compared to that of a higher GDP range of 1.0 to 1.5.


Let's now dive deeper into the life expectancy and GDP of each individual country.

# Life Expectancy Graphs

Has life expectancy increased overtime in the six nations?



```python
life_expectancy_graphs = sns.FacetGrid(data, col = "country", hue = "country", col_wrap = 3, sharey = False)
life_expectancy_graphs.map(sns.lineplot,  "year", "life_expectancy")
life_expectancy_graphs

                                                
```

Based on the six life expectancy graphs above, it is clear that life expectancy rose for all six countries from years 2000 to 2015, with an anomaly observed in the case of Zimbabwe where there was a noticeable dip in life expectancy from 2000 to 2004. 


Let's now take a closer look at how the GDP has changed for the different countries from 2000-2015. 

# GDP analysis

Has GDP increased overtime in the six nations?

```python
GDP_graphs = sns.FacetGrid(data, col = "country", hue = "country" ,col_wrap = 3, sharey = False)
GDP_graphs.map(sns.lineplot, "year", "GDP")
GDP_graphs
```

Based on the graphs above, we can see that the GDP generally rose for all six countries from 2000 to 2015. Again, there is an anomaly in the case of Zimbabwe whereby there was a noticeable decrease in GDP from 2000 to about 2008. 

Based on all the graphs shown above, there is strong evidence allowing us to infer that there is a positive correlation between life expectancy and GDP across all six countries, albeit some anomalies present in certain years, such as sudden dips in life expectancy and GDP for Zimbabwe. 

## What is the average life expectancy in these nations?  




```python
dfavg = data.groupby("country").mean().reset_index()
```

```python
sns.barplot(data = dfavg, x = "life_expectancy", y = "country" )
plt.xlabel("Life Expectancy (years)")
plt.ylabel("Country")
```

Based on the barplot above, Germany has the highest life expectancy, and the five countries (excluding Zimbabwe) have relatively similar life expectancies ranging from about mid-70s to low 80s, as opposed to Zimbabwe which has a much lower life expectancy of about 50. 

This noticeably low life expectancy in Zimbabwe also explains the left-skewed nature of the life expectancy distribution we saw earlier with the histogram plot, with Zimbabwe most likely contributing to the concentration of low life expectancies at the tail of the distribution.

## Is there a correlation between the life expectancy and GDP of a country? 

Now that we've done some analysis on both the life expectancy and GDP factors individually, let's explore the relationship between the two factors to conclude our analysis. 

```python
correlation_graphs = sns.FacetGrid(data, col = "country", hue = "country", col_wrap = 3, sharey = False, sharex = False)
correlation_graphs.map(sns.scatterplot, "GDP", "life_expectancy")
```

Based on the 6 scatterplots above, we can infer that there is a **positive** correlation between life expectancy and GDP in all of the 6 countries. In other words, the higher the GDP of a resident in a nation, the higher the projected life expectancy of that person.
